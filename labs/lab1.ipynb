```python
from bs4 import BeautifulSoup
import requests
```

```python
countries = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
response = requests.get(countries)
print(response.text[0:1000])
```

```python
soup = BeautifulSoup(response.text, "lxml")
title = soup.find('title')
print("TITLE", title)
```

```python
tables = soup.find_all('tr')
t = str(tables[2])
t
```

```python
text = [el.text for el in soup.find_all('td')]
text = text[8:]
text_split = [elem.split() for elem in text]
text_split[:20]
```

```python
china = ["China"]
china = list(china)
countriess = text_split[6::7]
countries = china + countriess

countries = countries[:240]
```

```python
china = ['1,409,670,000']
china = list(china)
populationss = text_split[7::7]
populations = china + populationss

populations = populations[:240]
```

```python
countries = ["".join(x) for x in countries]
```

```python
populations = ["".join(x) for x in populations]
for i, population in enumerate(populations):
    if "," in population:
        populations[i] = population.split(",")
```

```python
populations = ["".join(x) for x in populations]
```

```python
countries_populations = {}

for i in range(len(countries)):
    countries_populations[countries[i]] = populations[i]
countries_populations
```

```python
import pandas as pd
countries_populations = pd.DataFrame.from_dict(countries_populations, orient='index', columns=['Population'])
countries_populations
```

```python
countries_populations.to_csv('countries_populations.csv', index=True, header=False)
```
